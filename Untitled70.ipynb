```python
#1
import pandas as pd
# Load the dataset
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
df = pd.read_csv(url)
print(df.head())

# Check for missing values
print(df.isnull().sum())

# Fill missing values with linear interpolation
df['Passengers'] = df['Passengers'].interpolate(method='linear')

# Convert the 'Month' column to datetime
df['Month'] = pd.to_datetime(df['Month'])

# Set the 'Month' column as the index
df.set_index('Month', inplace=True)

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Passengers'], label='Passengers')
plt.xlabel('Time')
plt.ylabel('Number of Passengers')
plt.title('Monthly Number of Air Passengers')
plt.legend()
plt.show()

from statsmodels.tsa.seasonal import seasonal_decompose
# Decompose the time series
result = seasonal_decompose(df['Passengers'], model='multiplicative')
# Plot the decomposition
result.plot()
plt.show()



```

```python
#2.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.api import SimpleExpSmoothing
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
import warnings
warnings.filterwarnings('ignore')

url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master//airline-passengers.csv'
df = pd.read_csv(url)
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)
train, test = df[:'1959'], df['1960':]

ses_model = SimpleExpSmoothing(train['Passengers']).fit()
ses_forecast = ses_model.forecast(len(test))

sma_forecast = train['Passengers'].rolling(window=12).mean().iloc[-1]
sma_forecast = pd.Series([sma_forecast] * len(test), index=test.index)

hw_model = ExponentialSmoothing(train['Passengers'], seasonal='add', seasonal_periods=12).fit()
hw_forecast = hw_model.forecast(len(test))
def calculate_metrics(true, forecast):
    mae = mean_absolute_error(true, forecast)
    mse = mean_squared_error(true, forecast)
    rmse = np.sqrt(mse)
    mape = mean_absolute_percentage_error(true, forecast)
    return mae, mse, rmse, mape

ses_mae, ses_mse, ses_rmse, ses_mape = calculate_metrics(test['Passengers'], ses_forecast)
sma_mae, sma_mse, sma_rmse, sma_mape = calculate_metrics(test['Passengers'], sma_forecast)
hw_mae, hw_mse, hw_rmse, hw_mape = calculate_metrics(test['Passengers'], hw_forecast)


print(f'Simple Exponential Smoothing - MAE: {ses_mae}, MSE: {ses_mse}, RMSE: {ses_rmse}, MAPE: {ses_mape}')
print(f'Simple Moving Average - MAE: {sma_mae}, MSE: {sma_mse}, RMSE: {sma_rmse}, MAPE: {sma_mape}')
print(f'Holt-Winters Exponential Smoothing - MAE: {hw_mae}, MSE: {hw_mse}, RMSE: {hw_rmse}, MAPE: {hw_mape}')

decomposition = seasonal_decompose(df['Passengers'], model='additive', period=12)
plt.figure(figsize=(14, 8))
decomposition.plot()

plt.figure(figsize=(14, 8))
plt.plot(train.index, train['Passengers'], label='Train')
plt.plot(test.index, test['Passengers'], label='Test', color='gray')
plt.plot(test.index, ses_forecast, label='SES Forecast', color='blue')
plt.plot(test.index, sma_forecast, label='SMA Forecast', color='green')
plt.plot(test.index, hw_forecast, label='Holt-Winters Forecast', color='red')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.title('Forecasting with Different Techniques')
plt.legend(), plt.show()

```

```python
#3
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
df = pd.read_csv(url)


# Generate white noise data
np.random.seed(42)
white_noise = np.random.normal(loc=0.0, scale=1.0, size=len(df))

# Create a DataFrame for the white noise data
white_noise_df = pd.DataFrame({'Month': df['Month'], 'WhiteNoise': white_noise})

white_noise_df['Month'] = pd.to_datetime(white_noise_df['Month'])

white_noise_df.set_index('Month', inplace=True)
# Plot the white noise data
plt.figure(figsize=(10, 5))
plt.plot(white_noise_df.index, white_noise_df['WhiteNoise'], label='White Noise')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('White Noise Data')
plt.legend()
plt.show()

# Plot both the white noise and the time series data
plt.figure(figsize=(14, 7))
# Plot white noise
plt.subplot(2, 1, 1)
plt.plot(white_noise_df.index, white_noise_df['WhiteNoise'], label='White Noise')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('White Noise Data')
plt.legend()
# Plot Air Passenger time series data
plt.subplot(2, 1, 2)
plt.plot(df['Month'], df['Passengers'], label='Air Passengers')
plt.xlabel('Time')
plt.ylabel('Number of Passengers')
plt.title('Monthly Number of Air Passengers')
plt.legend()
plt.tight_layout()
plt.show()



from statsmodels.tsa.stattools import adfuller, kpss
# Augmented Dickey-Fuller Test
def adf_test(series):
  result = adfuller(series)
  print('ADF Statistic:', result[0])
  print('p-value:', result[1])
  for key, value in result[4].items():
    print('Critical Values:')
    print(f' {key}, {value}')

# KPSS Test
def kpss_test(series):
  result = kpss(series, regression='c')
  print('KPSS Statistic:', result[0])
  print('p-value:', result[1])
  for key, value in result[3].items():
    print('Critical Values:')
    print(f' {key}, {value}')

print("ADF Test for White Noise:")
adf_test(white_noise_df['WhiteNoise'])
print("\nKPSS Test for White Noise:")
kpss_test(white_noise_df['WhiteNoise'])
print("\nADF Test for Air Passengers:")
adf_test(df['Passengers'])
print("\nKPSS Test for Air Passengers:")
kpss_test(df['Passengers'])





```

```python
#5
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.ar_model import AutoReg
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
df = pd.read_csv(url)
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)

plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Passengers'], label='Passengers')
plt.xlabel('Time')
plt.ylabel('Number of Passengers')
plt.title('Monthly Number of Air Passengers')
plt.legend()
plt.show()

# Plot ACF
plt.figure(figsize=(10, 5))
plot_acf(df['Passengers'], lags=40, ax=plt.gca())
plt.xlabel('Lags')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function (ACF)')
plt.show()
# Plot PACF
plt.figure(figsize=(10, 5))
plot_pacf(df['Passengers'], lags=40, ax=plt.gca())
plt.xlabel('Lags')
plt.ylabel('Partial Autocorrelation')
plt.title('Partial Autocorrelation Function (PACF)')
plt.show()

# Fit AR(1) model
model_ar1 = AutoReg(df['Passengers'], lags=1).fit()
predictions_ar1 = model_ar1.predict(start=len(df), end=len(df)+11, dynamic=False)
# Plot AR(1) predictions
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Passengers'], label='Passengers')
plt.plot(predictions_ar1.index, predictions_ar1, label='AR(1) Predictions', color='red')
plt.xlabel('Time')
plt.ylabel('Number of Passengers')
plt.title('AR(1) Model Predictions')
plt.legend()
plt.show()
# Calculate evaluation metrics for AR(1)
ar1_mae = mean_absolute_error(df['Passengers'][-12:], predictions_ar1)
ar1_mse = mean_squared_error(df['Passengers'][-12:], predictions_ar1)
ar1_rmse = np.sqrt(ar1_mse)
print(f'AR(1) MAE: {ar1_mae:.2f}')
print(f'AR(1) MSE: {ar1_mse:.2f}')
print(f'AR(1) RMSE: {ar1_rmse:.2f}')

#explore higher lags AR
# Fit AR(2) model
model_ar2 = AutoReg(df['Passengers'], lags=2).fit()
predictions_ar2 = model_ar2.predict(start=len(df), end=len(df)+11, dynamic=False)
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Passengers'], label='Passengers')
plt.plot(predictions_ar2.index, predictions_ar2, label='AR(2) Predictions', color='green')
plt.xlabel('Time')
plt.ylabel('Number of Passengers')
plt.title('AR(2) Model Predictions')
plt.legend()
plt.show()
# Calculate evaluation metrics for AR(2)
ar2_mae = mean_absolute_error(df['Passengers'][-12:], predictions_ar2)
ar2_mse = mean_squared_error(df['Passengers'][-12:], predictions_ar2)
ar2_rmse = np.sqrt(ar2_mse)
print(f'AR(2) MAE: {ar2_mae:.2f}')
print(f'AR(2) MSE: {ar2_mse:.2f}')
print(f'AR(2) RMSE: {ar2_rmse:.2f}')
# Fit AR(3) model
model_ar3 = AutoReg(df['Passengers'], lags=3).fit()
predictions_ar3 = model_ar3.predict(start=len(df), end=len(df)+11, dynamic=False)
# Plot AR(3) predictions
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Passengers'], label='Passengers')
plt.plot(predictions_ar3.index, predictions_ar3, label='AR(3) Predictions', color='purple')
plt.xlabel('Time')
plt.ylabel('Number of Passengers')
plt.title('AR(3) Model Predictions')
plt.legend()
plt.show()
# Calculate evaluation metrics for AR(3)
ar3_mae = mean_absolute_error(df['Passengers'][-12:], predictions_ar3)
ar3_mse = mean_squared_error(df['Passengers'][-12:], predictions_ar3)
ar3_rmse = np.sqrt(ar3_mse)
print(f'AR(3) MAE: {ar3_mae:.2f}')
print(f'AR(3) MSE: {ar3_mse:.2f}')
print(f'AR(3) RMSE: {ar3_rmse:.2f}')









```
